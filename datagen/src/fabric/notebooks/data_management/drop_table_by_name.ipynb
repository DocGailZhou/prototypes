# Drop Specific Table by Name

This notebook allows you to drop (completely remove) specific tables by providing the schema name and table name.
- Completely removes the table and its data
- Validates schema and table existence before operation
- Provides clear feedback and lists available options if table not found

```python
# === CONFIGURE TABLE TO DROP ===
SCHEMA_NAME = "shared"      # Change this to your schema name
TABLE_NAME = "customer"     # Change this to your table name

# === DROP TABLE ===
def drop_table_by_name(schema_name, table_name):
    """Drop a specific table by schema and table name"""
    full_table_name = f"{schema_name}.{table_name}"
    
    try:
        # Get row count before dropping (optional info)
        try:
            row_count = spark.sql(f"SELECT COUNT(*) as count FROM {full_table_name}").collect()[0]['count']
            print(f"📊 Dropping table with {row_count:,} rows...")
        except:
            print(f"🗑️ Dropping table {full_table_name}...")
        
        # Drop the table
        spark.sql(f"DROP TABLE IF EXISTS {full_table_name}")
        print(f"✅ Successfully dropped {full_table_name}!")
        return True
        
    except Exception as e:
        print(f"❌ Failed to drop {full_table_name}: {e}")
        return False

# === EXECUTE ===
print(f"🎯 Target: {SCHEMA_NAME}.{TABLE_NAME}")
success = drop_table_by_name(SCHEMA_NAME, TABLE_NAME)

if success:
    print(f"🎉 Drop completed!")
else:
    print(f"💡 Check schema/table names and try again.")
```
