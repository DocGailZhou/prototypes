# Truncate Specific Table by Name

This notebook allows you to truncate (remove all data from) specific tables by providing the schema name and table name.
- Preserves table structure while removing all rows
- Validates schema and table existence before operation
- Provides clear feedback on the operation

```python
# === CONFIGURE TABLE TO TRUNCATE ===
SCHEMA_NAME = "schema_name"      # Change this to your schema name  
TABLE_NAME = "table_name"        # Change this to your table name

# === TRUNCATE TABLE ===
def truncate_table_by_name(schema_name, table_name):
    """Truncate a specific table by schema and table name"""
    full_table_name = f"{schema_name}.{table_name}"
    
    try:
        # Get row count before truncating (optional info)
        try:
            row_count = spark.sql(f"SELECT COUNT(*) as count FROM {full_table_name}").collect()[0]['count']
            print(f"📊 Truncating table with {row_count:,} rows...")
        except:
            print(f"🔨 Truncating table {full_table_name}...")
        
        # Truncate the table
        spark.sql(f"TRUNCATE TABLE {full_table_name}")
        print(f"✅ Successfully truncated {full_table_name}!")
        return True
        
    except Exception as e:
        print(f"❌ Failed to truncate {full_table_name}: {e}")
        return False

# === EXECUTE ===
print(f"🎯 Target: {SCHEMA_NAME}.{TABLE_NAME}")
success = truncate_table_by_name(SCHEMA_NAME, TABLE_NAME)

if success:
    print(f"🎉 Truncation completed! (Table structure preserved)")
else:
    print(f"💡 Check schema/table names and try again.")
```
